```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("xor.csv")
```

```python
data.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
data_array = np.array(data)
```

```python
x = data_array[:, 0:2]
y = data_array[:, -1]
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
def plot_decision_boundary(X, y, wt_in_hid, wt_hid_out, b_hid, b_out):
    # Prepare a grid of points in the range of input data
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    h = 0.01
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid_points = np.c_[xx.ravel(), yy.ravel()]

    # Make predictions for the grid of points
    hidden_layer_output = sigmoid(np.dot(grid_points, wt_in_hid) + b_hid)
    y_pred = sigmoid(np.dot(hidden_layer_output, wt_hid_out) + b_out)
    y_pred = (y_pred > 0.5).astype(int)  # Convert to binary predictions

    # Reshape the prediction results to the original grid shape
    y_pred = y_pred.reshape(xx.shape)

    # Display the decision boundary plot
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
    plt.xlabel('X_1')
    plt.ylabel('X_2')
    plt.title('Decision Boundary')
    plt.show()
```

```python
# Initialize KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

accuracies = []

# Iterate over the folds
for fold, (train_index, test_index) in enumerate(kf.split(x,y)):
    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]

    wt_in_hid = np.random.randn(2, 2)* np.sqrt(1 / 2)
    wt_hid_out = np.random.randn(2, 1)* np.sqrt(1 / 2)

    b_hid = np.zeros((1, 2))
    b_out = np.zeros((1, 1))

    lr = 0.1

    epochs=20000
    for epoch in range(epochs):

      output_hidden = sigmoid(np.dot(X_train, wt_in_hid) + b_hid)
      y_pred = sigmoid(np.dot(output_hidden, wt_hid_out) + b_out)

      # loss function
      error = y_train.reshape(-1, 1) - y_pred
      loss = np.mean(error**2)

      # Backpropagation
      d_output = error * sigmoid_derivative(y_pred)

      error_hidden = d_output.dot(wt_hid_out.T)
      d_hidden = error_hidden * sigmoid_derivative(output_hidden)

      # update weight and bias
      wt_hid_out += output_hidden.T.dot(d_output) * lr
      wt_in_hid += X_train.T.dot(d_hidden) * lr

      b_hid += np.sum(d_hidden, axis=0, keepdims=True) * lr
      b_out += np.sum(d_output, axis=0, keepdims=True) * lr

      # Test Data
      output_hidden_test = sigmoid(np.dot(X_test, wt_in_hid) + b_hid)
      y_pred_test = sigmoid(np.dot(output_hidden_test, wt_hid_out) + b_out)

      # Calculate accuracy
      y_pred_binary = (y_pred_test > 0.5).astype(int)
      accuracy = accuracy_score(y_test, y_pred_binary)
      accuracies.append(accuracy)

      # Print accuracy
      if epoch % 1000 == 0:
        print(f"Fold {fold+1}, Epoch {epoch}, Accuracy: {accuracy}")
        # print(f"y_pred {y_pred}, y_train {y_train}")

    print(f"Decision Boundary for Fold {fold + 1}")
    plot_decision_boundary(x, y, wt_in_hid, wt_hid_out, b_hid, b_out)
```

```python
# Average Accuracy
average_accuracy = np.mean(accuracies)
print(f"Average Accuracy: {average_accuracy}")
```
